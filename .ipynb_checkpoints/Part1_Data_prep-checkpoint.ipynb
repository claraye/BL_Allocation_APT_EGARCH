## Data Preparation: expected returns estimation using CAPM & APT (2 models)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
import pandas_datareader.data as web
import statsmodels.api as sm
from scipy import stats
from statsmodels.tsa.stattools import adfuller
import statsmodels.graphics.tsaplots as tsplt
from sklearn.decomposition import PCA
```

 - ###  1. _CAPM Model_
    - 1.1 First stage: get the mean of market data $R_{m}$
    - 1.2 Second stage: get regression coefficients $\beta_{i}$'s
    - 1.3 Estimate CAPM expected returns from CAPM model

#### 1.1 First stage: get the mean of market data $R_{m}$

```python
tickers = ['XRX','COTY','HES','CMG','XLNX','DVN','IPGP','ANET','AMD',
           'MSCI','HBI','NBL','CDNS','WDC','APC','KEYS','CELG','APTV',
           'CAG','QCOM','LRCX','SYF','ULTA','WYNN','APA','NVDA','NFLX',
           'PVH','SNPS','ALGN','FB','KLAC','TDG','AMP','BBY','MU','TSN',
           'GPN','CPRT','FLT','GRMN','DOV','MAS','MCHP','DISH','NSC',
           'JEC','HP','MCO','FAST']
n = 50
start_date = '2013-01-01'
end_date = '2018-12-31'

asset = web.DataReader(tickers, 'yahoo', start_date, end_date)
market = web.DataReader('^GSPC', 'yahoo', start_date, end_date)
risk_free = web.DataReader('^IRX', 'yahoo',start_date, end_date)

R_F = np.mean(risk_free['Adj Close']) / 252

ass_close = asset['Adj Close']
ass_close = ass_close.fillna(method='bfill')
ass_return = np.log(ass_close / ass_close.shift(1))
ass_return = ass_return.fillna(method='bfill')
ass_return = ass_return[1:]
ass_return *= 100
#close.describe()

market_close = market['Adj Close']
market_close = market_close.fillna(method='bfill')
mkt_return = np.log(market_close / market_close.shift(1))
mkt_return = mkt_return.fillna(method='bfill')
mkt_return = mkt_return[1:]
mkt_return *= 100
```

```python
#consider time series to predict market return
result = adfuller(mkt_return.values)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if result[1] <= 0.05:
    print('ADF test rejects H0. This series is stationary.')
else:
    print('ADF test cannot reject H0. This series might not be stationary.')
```

```python
#market return data can be regarded as weak stationary
fig = tsplt.plot_acf(mkt_return, alpha=.05, lags=24)
```

```python
#use historical data to calculate market return
mkt_mean = np.mean(mkt_return)
mkt_var = np.var(mkt_return)
```

#### 1.2 Second stage: get regression coefficients $\beta_{i}$'s

```python
ass_var = np.var(ass_return)
ass_cov = ass_return.cov()
mkt_ass_cov = np.dot(mkt_return.T - mkt_mean, ass_return - ass_return.mean(axis=0)) / (mkt_return.shape[0]-1)
# direct computation
beta = mkt_ass_cov / mkt_var

CAPM_return = R_F + beta * (mkt_mean - R_F)

excess_return = CAPM_return - R_F
excess_return_over_beta = excess_return / beta

#regression to get unsystematic error
y = ass_return - R_F
x = mkt_return - R_F
beta_reg = []
unsystematic_risk = []
for i in range(n):
    reg = sm.OLS(y.iloc[:,i], x.astype(float)).fit()
    beta_reg = np.append(beta_reg, reg.params)
    unsystematic_risk = np.append(unsystematic_risk, reg.bse)
data = pd.DataFrame({'Unsystematic Risk':unsystematic_risk})
data['Unsystematic Risk'] = data['Unsystematic Risk']**2
```

```python
data_CAPM = pd.DataFrame({'Expected Return':CAPM_return, 'Excess Return':excess_return,
                        'CAPM beta':beta, 'Regression beta':beta_reg, 'Unsystematic Risk':unsystematic_risk,
                        'Excess Return over beta':excess_return_over_beta})
data_CAPM['Unsystematic Risk'] = data['Unsystematic Risk']**2
data_CAPM.head()
```

 - ###  2. _APT - Classic Fama-French 3 Factors Model_
    - 2.1 First stage: get factor loadings $b_{ij}$'s
    - 2.2 Second stage: get $\lambda_j$'s
    - 2.3 Estimate APT expected returns from Fama-French 3 factors model

#### 2.1 First stage: get factor loadings $b_{ij}$'s

```python
# Adjusting asset returns to monthly frequency by taking the averages
ass_monthly_return = ass_return.groupby(pd.Grouper(freq='MS')).mean()

fama_data = web.DataReader('F-F_Research_Data_Factors', 'famafrench', start_date, end_date)
fama_factor = fama_data[0]
ass_monthly_return.index = fama_factor.index
R_RF = ass_monthly_return.sub(fama_factor['RF'], axis=0)

I = fama_factor[['Mkt-RF','SMB','HML']]
bij = np.empty(shape=(50,3))
for i in range(n):
    reg = sm.OLS(R_RF.iloc[:,i], I).fit()
    bij[i] = reg.params.T
R_RF_T = R_RF.T
bij_frame = pd.DataFrame(bij, columns=['b1', 'b2', 'b3'])
bij_frame.index = R_RF_T.index
bij_frame.head()
```

#### 2.2 Second stage: get $\lambda_j$'s

```python
m = len(R_RF.index)
lambdatj = np.empty(shape=(m,3))
for t in range(m):
    cross_reg = sm.OLS(R_RF_T.iloc[:,t], bij_frame).fit()
    lambdatj[t] = cross_reg.params.T
lambdatj = pd.DataFrame(lambdatj, columns=['lambda1', 'lambda2', 'lambda3'])
lambdaj = np.mean(lambdatj)
lambdaj
```

#### 2.3 Estimate APT expected returns from Fama-French 3 factors model

```python
#APT Fama FrenchModel
R_bar_fama = R_F + np.matmul(bij_frame.values, lambdaj.T)
R_bar_fama = pd.DataFrame(R_bar_fama, index=R_RF.columns, columns=['R_bar_fama'])
R_bar_fama.T
```

 - ###  3. _APT - Proposed Multi-factors Model_
    - 3.1 Load original factor values $I_j$'s (from the file _multi-factors.csv_)
    - 3.2 Conduct PCA to determine factors $I^*_j$'s
    - 3.3 First stage: get factor loadings $b_{ij}$'s
    - 3.4 Second stage: get $\lambda_j$'s
    - 3.5 Estimate APT expected returns from multi-factors model

#### 3.1 Load original factor values $I_j$'s

```python
# Load factors data I_j
factors = pd.read_csv("./multi-factors.csv")
factors = factors.set_index('month')
factors.index = ass_monthly_return.index
R_RF = ass_monthly_return.sub(factors['RF'], axis=0)
factors = factors.drop(columns='RF')
factors.describe()
```

```python
print("Number of factors to select from: %d" % len(factors.columns))
```

#### 3.2 Conduct PCA to determine factors $I^*_j$'s

```python
# Standardized the factors
standard_factors=(factors - factors.mean())/factors.std()
standard_factors.describe()
```

Covariance matrix before PCA

```python
# Calculate the covariance matrix of factors before PCA
factors_cov = standard_factors.cov()
factors_cov
```

```python
# Use PCA to transfrom the standardized factor dataset.
pca = PCA(n_components=0.95)   # to explain 95% variations
factors_pca_mat = pca.fit_transform(standard_factors.values)
num_of_selected = factors_pca_mat.shape[1]
print("Number of factors after PCA: %d" % num_of_selected)
```

```python
# Calculate the covariance matrix of factors after PCA
column_names = ['f%d' %(i+1) for i in range(num_of_selected)]
factors_pca_df = pd.DataFrame(factors_pca_mat, columns=column_names)
factors_pca_df.index = ass_monthly_return.index
```

Covariance matrix after PCA

```python
factors_pca_cov = factors_pca_df.cov()
factors_pca_cov
```

```python
# The final factors data used for APT
I = factors_pca_df
I.head()
```

#### 3.3 First stage: get factor loadings $b_{ij}$'s   (by time series regressions of each stocks)

```python
bij = np.empty(shape=(n, num_of_selected))
for i in range(n):
    reg = sm.OLS(R_RF.iloc[:,i], I).fit()
    bij[i] = reg.params.T
bij_column_names = ['b%d' %(i+1) for i in range(num_of_selected)]
bij_df = pd.DataFrame(bij, columns=bij_column_names, index=R_RF.columns)
bij_df.head()
```

#### 3.4 Second stage: get $\lambda_j$'s
##### (a) get $\lambda_{t, j}$'s by cross-sectional regressions on each month

```python
lambdatj = np.empty(shape=(len(R_RF.index), num_of_selected))
for t in range(len(R_RF.index)):
    cross_reg = sm.OLS(R_RF.iloc[t,:], bij_df).fit()
    lambdatj[t] = cross_reg.params.T
lambdatj_column_names = ['lambda%d' %(i+1) for i in range(num_of_selected)]
lambdatj = pd.DataFrame(lambdatj, columns=lambdatj_column_names, index=R_RF.index)
lambdatj.head()
```

##### (b) get $\lambda_j$'s by taking the average of $\lambda_{t,j}$'s

```python
lambdaj = np.mean(lambdatj)
lambdaj
```

#### 3.5 Estimate APT expected returns from multi-factors model

```python
R_bar_multi = R_F + bij_df.values.dot(lambdaj)
R_bar_multi = pd.DataFrame(R_bar_multi, index=R_RF.columns, columns=['R_bar_multi'])
R_bar_multi.T
```

## Data Comparison: expected returns estimation using CAPM & APT (2 models)

```python
names = ass_return.columns.tolist()

fig, ax = plt.subplots(figsize=(30,6))
ax.scatter(names, CAPM_return, color='black', marker='P', label='CAPM')
ax.scatter(names, R_bar_fama, color='red', marker='o', label='APT-Fama French')
ax.scatter(names, R_bar_multi, color='blue', marker='^', label='APT-Multi Index')

legend = ax.legend(loc=0, fontsize='x-large')

plt.title('Expected Return Plot',fontsize=24)
```
